```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
from pmdarima import auto_arima
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
import math
import pickle
```

```python
source = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/country_data/South%20Africa.csv"
df = pd.read_csv(source,
                parse_dates=True)
df1=df[df["date"]> "2021-07-06"]
df3 = df1[df1["people_fully_vaccinated"].notnull()]
df3
```

```python
df = df3[["date","people_fully_vaccinated"]]
df=df.reset_index()
df
```

```python
vv = -2
x_train = df.iloc[:vv]
x_test = df.iloc[vv:]
x_train.shape
target = "people_fully_vaccinated"
```

```python
#Get order
s_fit=auto_arima(df[target],trace=False)
s_fit.summary()
```

```python
model = ARIMA(x_train["people_fully_vaccinated"], order=(0,1,0))
model=model.fit()
```

```python
startV = len(x_train)
endV = len(x_train)+len(x_test)
my_pred=model.predict(startV,endV-1,typ="levels")
my_pred
```

```python
#ASSESS THE MODEL
m1 = df["people_fully_vaccinated"].mean()
m2 = math.sqrt(mean_squared_error(my_pred,df[target][startV-1:endV-1]))
print(m1, m2)
```

```python
# Save model
name = "model_"+target
with open(name, 'wb') as f:
    pickle.dump(model,f)
```

```python
# Save values
vname = "values_"+target+".txt"
strV = f'{startV} {endV}'
with open(vname,'w') as f:
        f.write(strV)
```
